# KERALA HOSPITAL TREATMENTS

### The dataset titled "Kerala Hospital Treatments" contains detailed synthetic information about hospitals and treatments in Kerala, India. It comprises 15 columns and 1000 rows, providing comprehensive data related to healthcare in the region. Each record includes unique identifiers for hospitals, patients, and treatments, along with the hospital name, location (major cities and districts in Kerala such as Thiruvananthapuram, Ernakulam, and Kozhikode), and the hospital's medical specialty (e.g., Cardiology, Pediatrics, Neurology).

The dataset also includes patient demographics such as age and gender, along with information about their diagnosis (e.g., Diabetes, Hypertension, Cancer) and the type of treatment received (e.g., Surgery, Medication, Therapy). Financial details like the treatment cost in INR and insurance coverage status are also included. Dates of admission and discharge provide a timeline for treatment, while the outcome column indicates whether the patient recovered, improved, or showed no change. This dataset is ideal for analysis of healthcare trends, cost evaluation, and patient demographics in Kerala's medical institutions.

#### 1.How many unique hospitals are present in the dataset, and what are their specialties ?

```python

unique_hospitals = df[['Hospital_ID', 'Specialty']].drop_duplicates()

num_unique_hospitals = unique_hospitals['Hospital_ID'].nunique()

specialties_per_hospital = unique_hospitals.groupby('Hospital_ID')['Specialty'].unique()

print(f"Number of unique hospitals: {num_unique_hospitals}")
print("Specialties for each hospital:")
print(specialties_per_hospital)
```

#### 2.What is the distribution of patients across different locations in Kerala?

```python

patient_distribution = df['Location'].value_counts()

print("Distribution of patients across locations:")
print(patient_distribution)
```


#### 3.What are the most common types of treatments provided in the hospitals?

```python

treatment_distribution = df['Treatment_Type'].value_counts()

print("Most common types of treatments provided in the hospitals:")
print(treatment_distribution)
```

#### 4.Which diagnoses are most frequently treated across all hospitals?

```python

diagnosis_frequency = df['Diagnosis'].value_counts()

print("Most frequently treated diagnoses across all hospitals:")
print(diagnosis_frequency)
```

#### 5.What is the average, minimum, and maximum cost of treatments in the dataset?

```python

average_cost = df['Treatment_Cost'].mean()
min_cost = df['Treatment_Cost'].min()
max_cost = df['Treatment_Cost'].max()

print(f"Average treatment cost: ₹{average_cost:.2f}")
print(f"Minimum treatment cost: ₹{min_cost:.2f}")
print(f"Maximum treatment cost: ₹{max_cost:.2f}")
```

#### 6.How many treatments were covered by insurance, and how many were not?

```python

insurance_coverage_counts = df['Insurance_Coverage'].value_counts()

print("Number of treatments covered by insurance:")
print(insurance_coverage_counts)
```

#### 7.What is the average length of hospital stay (difference between discharge and admission dates)?

```python

df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])
df['Discharge_Date'] = pd.to_datetime(df['Discharge_Date'])

df['Length_of_Stay'] = (df['Discharge_Date'] - df['Admission_Date']).dt.days

average_length_of_stay = df['Length_of_Stay'].mean()

print(f"The average length of hospital stay is: {average_length_of_stay:.2f} days.")
```

#### 8.How has the number of admissions changed over time (e.g., monthly trends)?

```python

df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])

df['Year_Month'] = df['Admission_Date'].dt.to_period('M')

monthly_admissions = df['Year_Month'].value_counts().sort_index()

print("Monthly trends in the number of admissions:")
print(monthly_admissions)
```

#### 9.What percentage of patients recovered, improved, or showed no change after treatment?

```python

outcome_percentages = df['Outcome'].value_counts(normalize=True) * 100

print("Percentage of patients by treatment outcome:")
print(outcome_percentages)
```

#### 10.Are certain specialties associated with better outcomes compared to others?

```python

outcome_by_specialty = df.groupby('Specialty')['Outcome'].value_counts(normalize=True).unstack() * 100

print("Outcome distribution by specialty (in percentage):")
print(outcome_by_specialty)
```

#### 11.Which city or district has the highest number of treatments performed?

```python

treatment_counts_by_location = df['Location'].value_counts()

top_location = treatment_counts_by_location.idxmax()
top_location_count = treatment_counts_by_location.max()

print(f"The city or district with the highest number of treatments is {top_location} with {top_location_count} treatments.")
```

#### 12.What is the average treatment cost for each city or district?

```python

average_cost_by_location = df.groupby('Location')['Treatment_Cost'].mean()

print("Average treatment cost for each city or district:")
print(average_cost_by_location)
```

#### 13.Is there a correlation between the type of treatment, diagnosis, or patient demographics (age and gender) and treatment outcomes?

```python

treatment_outcome_correlation = pd.crosstab(df['Treatment_Type'], df['Outcome'], normalize='index') * 100

diagnosis_outcome_correlation = pd.crosstab(df['Diagnosis'], df['Outcome'], normalize='index') * 100


gender_outcome_correlation = pd.crosstab(df['Patient_Gender'], df['Outcome'], normalize='index') * 100

age_outcome_correlation = df.groupby('Outcome')['Patient_Age'].mean()

print("Correlation between treatment type and outcome (in percentage):")
print(treatment_outcome_correlation)

print("\nCorrelation between diagnosis and outcome (in percentage):")
print(diagnosis_outcome_correlation)

print("\nGender and outcome correlation (in percentage):")
print(gender_outcome_correlation)

print("\nAverage age for each outcome:")
print(age_outcome_correlation)
```

#### 14. What is the distribution of treatment costs across different locations (cities or districts)?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(x='Location', y='Treatment_Cost', data=df)
plt.xticks(rotation=90)
plt.title('Treatment Cost Distribution Across Different Locations')
plt.show()
```

#### 15.How do treatment outcomes vary across different specialties?

```python

outcome_by_specialty = df.groupby('Specialty')['Outcome'].value_counts().unstack().fillna(0)
outcome_by_specialty = outcome_by_specialty.div(outcome_by_specialty.sum(axis=1), axis=0) * 100

outcome_by_specialty.plot(kind='bar', stacked=True, figsize=(12, 6), cmap='Set2')
plt.title('Treatment Outcomes by Specialty')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.show()
```
